```python
import matplotlib.pyplot as plt 

# problem 2a
x = [i/100 for i in range(-1000,1000)]
y = [num**2 for num in x]

plt.plot(x,y, color = "grey", linewidth = 2)
plt.grid(color='black', linestyle='-', linewidth=1)
plt.tick_params(
    axis = 'both',
    labelbottom = False,
    labelleft = False
)
```

```python
# four turns even degree
f = lambda u: (u - 2)*(u - 1)*(u + 1)*(u + 2)
x = [i/100 for i in range(-250,250)]
y = [f(i) for i in x]

plt.plot(x,y, color = "grey", linewidth = 2)
plt.grid(color = "black", linewidth = 1, linestyle = "-")
plt.tick_params(
    axis = "both",
    labelbottom = False,
    labelleft = False
)
```

```python
# roller coaster function

x = [i/100 for i in range(-425,210)]
y = [0.5*i*(i - 2)*(i - 1)*(i + 3)*(i + 4) for i in x]

plt.plot(x,y, color = "grey", linewidth = 2)
plt.grid(color = "black", linewidth = 1, linestyle = "-")
plt.tick_params(
    axis = "both",
    labelbottom = False,
    labelleft = False
)
```

```python
### Task 47 
from random import random
from random import seed
import matplotlib.pyplot as plt
seed(10)

func = lambda x: -(x - 18)**4 + 2000 + 10*random()*(240 - x) + random()*x

hour = [i/10 for i in range(1, 241)]
traffic = [func(i) for i in hour]

for i in range(len(traffic)):
    if traffic[i] < 0:
        traffic[i] = 0 + abs(200*random())
    elif traffic[i] > 3650:
        traffic[i] = 3650 
d = traffic

plt.scatter(hour, traffic, color = "black")
plt.title("Web Traffic vs Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Clients")
```

```python
## PW - Linear Model

func1 = lambda x: 0
func2 = lambda x: 90*(x - 100)
func3 = lambda x: 3650
func4 = lambda x: 3650 - 75*(x - 190)

hour = [i/10 for i in range(1, 241)]
traffic = [0 for i in hour]

for i in range(len(traffic)):
    if i <= 100:
        traffic[i] = func1(i)
    elif i <= 140:
        traffic[i] = func2(i)
    elif i <= 190:
        traffic[i] = func3(i)
    else:
        traffic[i] = func4(i)

A = traffic

plt.plot(hour, traffic, color = "black")
plt.title("Web Traffic vs Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Clients")
```

```python
## PW - Quad

func1 = lambda x: 0
func2 = lambda x: 2.2*(x - 100)**2
func3 = lambda x: 3650
func4 = lambda x: 3650 - 1.5*(x - 190)**2

hour = [i/10 for i in range(1, 241)]
traffic = [0 for i in hour]

for i in range(len(traffic)):
    if i <= 100:
        traffic[i] = func1(i)
    elif i <= 140:
        traffic[i] = func2(i)
    elif i <= 190:
        traffic[i] = func3(i)
    else:
        traffic[i] = func4(i)

B = traffic

plt.plot(hour, traffic, color = "black")
plt.title("Web Traffic vs Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Clients")
```
