```python
### Task 46 ###
import matplotlib.pyplot as plt 
from random import *
seed(11)
rpm = [i for i in range(5501)]
rpm_sample = [rpm[randint(500, 5500)] for i in range(100)]
sedan = [(10**-9)*x**2 + 10**-2*random() for x in rpm_sample]
truck = [(10**-16)*x**4 + 10**-2*random() + 0.01 for x in rpm_sample]
h_minivan = [2*(10**-9)*x**2 + 10**-2*random() for x in rpm_sample]
for i, v in enumerate(rpm_sample):
    if v <= 1500:
        h_minivan[i] = 0

plt.scatter(rpm_sample, truck, color = "red")
plt.scatter(rpm_sample, sedan, color = "blue", marker = "+")
plt.scatter(rpm_sample, h_minivan, color = "green", marker = "x")
plt.title("Fuel Consumption vs RPM by Vehicle Type")
plt.xlabel("RPM")
plt.ylabel("Fuel Consumption (L/min)") 
plt.legend(("Truck", "Sedan", "Hybrid Minivan"))
plt.xlim(0,6000)
```

```python
### Task 23
import matplotlib.pyplot as plt
 
func_c1 = lambda x: (x + 10)/x
func_c2 = lambda x: (10*x + 2)/x
func_c3 = lambda x: (x + 20)/x

nodes = [i for i in range(1,17)]
c1 = [func_c1(i) for i in nodes]
c2 = [func_c2(i) for i in nodes]
c3 = [func_c3(i) for i in nodes]

plt.scatter(nodes, c1, marker = "*")
plt.scatter(nodes, c2, marker = "+")
plt.scatter(nodes, c3)
plt.title("Average Time vs Number of Nodes")
plt.xlabel("Number of Nodes")
plt.ylabel("Average Time to Complete Job (min)")
plt.legend(("Computer A", "Computer B", "Computer C"))
```
